# Answer 1


```python
# import webdriver for scrap data 
driver = webdriver.Chrome()

# now call the url for scraping
driver.get('https://www.naukri.com/')
time.sleep(10)

# find the designation word in search bar 
find_des = driver.find_element(By.CLASS_NAME,'suggestor-input')
find_des.send_keys('Data Scientist')
time.sleep(5)

# find out the job location Delhi/NCR
find_loc = driver.find_element(By.XPATH,'//input[@placeholder="Enter location"]')
find_loc.send_keys('Delhi/NCR')
time.sleep(4)

# now to going to search the both input 
serc_butt = driver.find_element(By.CLASS_NAME,'qsbSubmit')
serc_butt.click()
time.sleep(5)

# find the 10 title names 
find_title = []
title = driver.find_elements(By.XPATH,'//a[@class="title "]')
for i in title[:10]:
    find = i.text.strip()  # Remove any leading/trailing spaces
    if not find:  # If title is empty or blank
        find = 'NA Title'
    find_title.append(find)
find_title

# find the 10 locations
find_loc = []
loc = driver.find_elements(By.XPATH,'//span[@class="ni-job-tuple-icon ni-job-tuple-icon-srp-location loc"]')
for i in loc[:10]:
    find = i.text
    find_loc.append(find)
find_loc

# find the 10 company names
find_com = []
com = driver.find_elements(By.XPATH,'//a[@class=" comp-name mw-25"]')
for i in com[:10]:
    find = i.text
    find_com.append(find)
find_com

# find the 10 experience requirements
job_exp = []
exp = driver.find_elements(By.XPATH,'//span[@class="ni-job-tuple-icon ni-job-tuple-icon-srp-experience exp"]')
for i in exp[:10]:
    find=i.text
    job_exp.append(find)
job_exp

# dataframe to show the result 
dict = { 'find_title' : find_title ,'find_loc' : find_loc , 
        'find_com' : find_com , 'job_exp' : job_exp}

df = pd.DataFrame(dict)
df
```

```python
# import webdriver and open Chrome browser
driver = webdriver.Chrome()

# navigate to the Shine website
driver.get('https://www.shine.com/')
time.sleep(5)

# find and click on the search input field
clic_butn = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.CLASS_NAME,'input')))
clic_butn.click()
time.sleep(5)

# find the job title input field and enter "Data Scientist"
serc_job = driver.find_element(By.ID,"id_q")
serc_job.send_keys('Data Scientist')

# find the location input field and enter "Bangalore"
serc_loc = driver.find_element(By.ID,"id_loc")
serc_loc.send_keys('Banglore')

# click on the search button
serc_buttn = driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[2]/div/button")
serc_buttn.click()

# create empty lists for job titles, locations, company names, and experience requirements
serc_job_title = []
serc_loc = []
comp_name = []
exper_req = []

# find and store the job titles in a list
fin_title = WebDriverWait(driver,10).until(EC.presence_of_all_elements_located((By.XPATH,'//strong[@class="jobCard_pReplaceH2__xWmHg"]')))
for title in fin_title[:10]:
    serc_job_title.append(title.text)

# find and store the locations in a list, removing unnecessary text like '\n+number'
loc = WebDriverWait(driver,10).until(EC.presence_of_all_elements_located((By.XPATH,'//div[@class="jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')))
for loca in loc[:10]:
    find_str = loca.text
    find = re.sub(r'\n\+\d','',find_str)
    serc_loc.append(find)

# find and store the company names in a list
fin_com_name = WebDriverWait(driver,10).until(EC.presence_of_all_elements_located((By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]')))
for title in fin_com_name[:10]:
    comp_name.append(title.text)

# find and store the experience requirements in a list
exper_ = WebDriverWait(driver,10).until(EC.presence_of_all_elements_located((By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')))
for title in exper_[:10]:
    exper_req.append(title.text)

# create a dictionary to organize the scraped data
dict = {'serc_job_title' : serc_job_title , 'comp_name' : comp_name , 'serc_loc' : serc_loc , 'exper_req' : exper_req}

# create a pandas DataFrame to display the data
df = pd.DataFrame(dict)
df
```

```python
# Open the Chrome browser using Selenium
driver = webdriver.Chrome()

# Navigate to Flipkart website
driver.get('https://www.flipkart.com/')
time.sleep(10)

# Wait for the search bar to be available and search for sneakers
serc_prod = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'Pke_EE')))
serc_prod.send_keys('sneakers')

# Wait for the search button to be clickable and click it
serc_butt = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CLASS_NAME, '_2iLD__')))
serc_butt.click()
time.sleep(3)

# Initialize empty lists to store brand names, descriptions, and prices
brand_name = []
brand_dsrcp = []
pro_pric = []

# Define start and end range for page navigation
start = 0
end = 3

# Loop through the pages to scrape brand names, descriptions, and prices
for page in range(start, end):
    # Wait for brand names to be visible and collect them
    br_name = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, '//div[@class="syl9yP"]')))
    for i in br_name[:100]:
        brand_name.append(i.text)
        if len(brand_name) == 100:
            break
    
    # Wait for product descriptions to be visible and collect them
    br_des_ = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, '//a[@class="WKTcLC BwBZTg" or @class="WKTcLC"]')))
    for i in br_des_[:100]:
        brand_dsrcp.append(i.text)
        if len(brand_dsrcp) == 100:
            break
    
    # Wait for product prices to be visible and collect them
    price = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, '//div[@class="Nx9bqj"]')))
    for i in price[:100]:
        pro_pric.append(i.text)
        if len(pro_pric) == 100:
            break
    
    # Click on the next button to navigate to the next page
    nxt_buttn = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CLASS_NAME, "_9QVEpD")))
    nxt_buttn.click()
    time.sleep(5)

# Create a dictionary of the scraped data
dict = {'brand_name': brand_name, 'pro_pric': pro_pric, 'brand_dsrcp': brand_dsrcp}

# Convert the dictionary into a pandas DataFrame and display it
df = pd.DataFrame(data=dict)
df
```

```python
# Initialize the Chrome WebDriver
driver = webdriver.Chrome()

# Navigate to the Jagran Josh website containing the list of Prime Ministers of India
driver.get('https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1')

# Initialize a list to store the names of the Prime Ministers
name = []
# Find all elements containing the names and append them to the list, skipping the header
quote = driver.find_elements(By.XPATH, '//div[@class="TableData"]//tr//td[2]')
for i in quote[1:-1]:
    name.append(i.text)

# Initialize a list to store the birth dates of the Prime Ministers
born_de = []
# Find all elements containing the birth dates and append them to the list
quote = driver.find_elements(By.XPATH, '//div[@class="TableData"]//tr//td[3]')
for i in quote[1:]:
    born_de.append(i.text)

# Initialize a list to store the terms of office of the Prime Ministers
trm_offc = []
# Find all elements containing the terms of office and append them to the list
quote = driver.find_elements(By.XPATH, '//div[@class="TableData"]//tr//td[4]')
for i in quote[1:]:
    trm_offc.append(i.text)

# Initialize a list to store remarks about the Prime Ministers
remrk = []
# Find all elements containing remarks and append them to the list
quote = driver.find_elements(By.XPATH, '//div[@class="TableData"]//tr//td[5]')
for i in quote[1:]:
    remrk.append(i.text)

# Create a dictionary with names, birth dates, terms of office, and remarks
dict = {'name': name, 'born_de': born_de, 'trm_offc': trm_offc, 'remrk': remrk}

# Convert the dictionary into a pandas DataFrame and display it
df = pd.DataFrame(dict)
df
```

```python
# Initialize the Chrome WebDriver and navigate to the Motor1 website
driver = webdriver.Chrome()

# Wait for the page to load fully before proceeding
driver.get('https://www.motor1.com/')
time.sleep(5)

# Locate the search input box and enter the search query for "50 most expensive cars"
serc_buttn = driver.find_element(By.XPATH, '/html/body/div[9]/div[2]/div/div/div[3]/div/div/div/form/input')
serc_buttn.send_keys('50 most expensive cars')
time.sleep(3)

# Locate and click the search button to initiate the search
click_butn = driver.find_element(By.XPATH, "/html/body/div[9]/div[2]/div/div/div[3]/div/div/div/form/button[1]")
click_butn.click()
time.sleep(2)

# Wait for the specific link to be clickable and then click it to view the results
click_butn2 = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[9]/div[9]/div/div[1]/div/div/div[1]/a")))
click_butn2.click()

# Initialize a list to store car names
Car_name = []

# Locate the car names and store them in the list, excluding the last element
name = driver.find_elements(By.XPATH, '//h3[@class="subheader"]')
for i in name[:-1]:
    Car_name.append(i.text)
    time.sleep(5)

# Initialize a list to store car prices
Car_prices = []

# Locate the price elements and extract their text
price_elements = driver.find_elements(By.XPATH, '/html/body/div[9]/div[7]/div[2]/div[1]/div[2]/div[2]//p//strong')
for element in price_elements:
    text = element.text
    # Use regex to find and format the prices (in million or billion)
    prices = re.findall(r'(\$\d[\d,]*(?:\.\d+)?\s*(million|billion)(?:\s*or\s)?)', text, re.IGNORECASE)
    formatted_prices = [match[0] for match in prices]
    Car_prices.append(formatted_prices)

# Create a dictionary to hold the car names and prices, then convert it to a DataFrame
dict = {'Car_name': Car_name, 'Car_prices': Car_prices}
df = pd.DataFrame(dict)
df  # Display the DataFrame
```
