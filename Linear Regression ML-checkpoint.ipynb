```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pickle
from sklearn.linear_model import LinearRegression 
lm=LinearRegression()
from sklearn.linear_model import Lasso,Ridge,LassoCV,RidgeCV
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error , mean_squared_error
import seaborn as sns
```

```python
#let load the dataset
df = pd.read_csv('https://raw.githubusercontent.com/training-ml/Files/refs/heads/main/Admission_Prediction.csv')
#let's check head
df.head()
```

```python
# Display summary statistics for each numerical column,
# including count, mean, standard deviation, minimum, 
# 25th percentile, median, 75th percentile, and maximum values
df.describe()
```

```python
#treat the null value by using fillna or we can use impute technic also 
df['GRE Score'] =df['GRE Score'].fillna(df['GRE Score'].mean()) 
df['TOEFL Score'] =df['TOEFL Score'].fillna(df['TOEFL Score'].mean()) 
df['University Rating'] =df['University Rating'].fillna(df['University Rating'].mode()[0]) 
```

```python
#describe again data 
df.describe()
```

```python
#drop column which is no relation with target
df.drop('Serial No.',axis=1,inplace=True)
```

```python
#check data set
df
```

```python
# Plot KDE (Kernel Density Estimate) for each column in the DataFrame
# Set figure size to ensure all plots are clearly visible
plt.figure(figsize=(20,15)) 

# Initialize plot counter
plotnumber = 1

# Loop through each column in the DataFrame
for column in df:
    # Plot up to 8 columns in a 2x4 grid
    if plotnumber <= 8:
        ax = plt.subplot(2, 4, plotnumber)  # Create subplot in a 2x4 layout
        sns.kdeplot(df[column])  # Plot KDE for the current column
        plt.xlabel(column, fontsize=20)  # Set x-axis label with a larger font size
    plotnumber += 1  # Move to the next subplot

# Display all KDE plots
plt.show()
```

```python
# Separate the features (X) and target variable (y)
# Drop the 'Chance of Admit' column from X, as it is the target variable
x = df.drop(columns=['Chance of Admit'])  # Features
y = df['Chance of Admit']  # Target variable
```

```python
# Plot scatter plots for each feature against the target variable 'Chance of Admit'
# Set figure size and background color
plt.figure(figsize=(20,15), facecolor='blue')

# Initialize plot counter
plotnumber = 1

# Loop through each feature in X
for column in x:
    # Plot up to 8 features in a 2x4 grid
    if plotnumber <= 8:
        ax = plt.subplot(2, 4, plotnumber)  # Create subplot in a 2x4 layout
        plt.scatter(x[column], y)  # Scatter plot of feature vs. target
        plt.xlabel(column, fontsize=20)  # Set x-axis label with a larger font size
        plt.ylabel('Chance of Admit', fontsize=20)  # Set y-axis label with a larger font size
    plotnumber += 1  # Move to the next subplot

# Adjust layout to prevent overlap between subplots
plt.tight_layout()
```

```python
# Standardize the feature data
# Initialize StandardScaler to scale features to have mean=0 and standard deviation=1
scaler = StandardScaler()

# Fit the scaler to X and transform it to create a standardized version of X
x_scaled = scaler.fit_transform(x)
```

```python
# Split the dataset into training and testing sets
# Use 25% of the data for testing and 75% for training
# Set a random state for reproducibility
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=.25,random_state=348)
```

```python
# Train the model on the training data
lm.fit(x_train,y_train)
```

```python
# Display the last two rows of the DataFrame
df.tail(2)
```

```python
# Predict the chance of admission for a new data point
# First, scale the new input data to match the training data's scaling
# Then, use the trained model to predict the admission chance
print('admission of chance......_.',lm.predict(scaler.transform([[327.0,113.0,4.0,4.5,4.5,9.04,0]])))
```

```python
# Evaluate the model's performance on the training data
# The score method returns the R^2 (coefficient of determination) of the model on the training set
lm.score(x_train,y_train)
```

```python
# Evaluate the model's performance on the test data
# The score method returns the R^2 (coefficient of determination) of the model on the test set
lm.score(x_test,y_test)
```

```python
# Use the trained model to make predictions on the test data
# Predict the target values (Chance of Admit) for the test set features (x_test)
y_pred = lm.predict(x_test)
```

```python
# Create a scatter plot to visualize the relationship between actual and predicted values
# The x-axis represents the actual values (y_test), and the y-axis represents the predicted values (y_pred)
plt.scatter(y_test, y_pred)

# Label the axes for clarity
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')

# Display the plot
plt.show()
```

```python
# Calculate the Mean Absolute Error (MAE) between the actual and predicted values
# MAE measures the average magnitude of errors in the predictions, without considering their direction (positive/negative)
mean_absolute_error(y_test, y_pred)
```

```python
# Calculate the Mean Squared Error (MSE) between the actual and predicted values
# MSE measures the average of the squared differences between predicted and actual values
mean_squared_error(y_test, y_pred)
```

```python
# Calculate the Root Mean Squared Error (RMSE) between the actual and predicted values
# RMSE is the square root of the MSE, providing a more interpretable metric in the same units as the target variable
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
# Initialize the LassoCV model with cross-validation to select the optimal alpha
# 'alphas=None' allows LassoCV to automatically choose the best alpha value
# 'max_iter=10' sets the maximum number of iterations for the optimization process
Lasso_cv = LassoCV(alphas=None, max_iter=10)

# Fit the LassoCV model to the training data
Lasso_cv.fit(x_train, y_train)
```

```python
# Retrieve the optimal alpha value selected by LassoCV after fitting the model
alpha = Lasso_cv.alpha_

# Display the optimal alpha value
alpha
```

```python
# Initialize the Lasso regression model with the optimal alpha value obtained from LassoCV
lass = Lasso(alpha)

# Fit the Lasso regression model to the training data
lass.fit(x_train, y_train)
```

```python
# Evaluate the performance of the Lasso regression model on the test data
# The score method returns the R^2 (coefficient of determination) of the model on the test set
lass.score(x_test, y_test)
```

```python
# Generate a range of alpha values for Ridge regression with a step size of 0.01
# np.arange(0.001, 0.1, 0.01) creates an array of alphas from 0.001 to 0.099

# Initialize the RidgeCV model with cross-validation to select the optimal alpha from the generated range
rig = RidgeCV(alphas=np.arange(0.001, 0.1, 0.01))

# Fit the RidgeCV model to the training data
rig.fit(x_train, y_train)
```

```python
# Retrieve the optimal alpha value selected by RidgeCV after fitting the model
rig.alpha_
```

```python
# Initialize the Ridge regression model using the optimal alpha value from RidgeCV
rige = Ridge(alpha=rig.alpha_)

# Fit the Ridge regression model to the training data
rige.fit(x_train, y_train)
```

```python
# Evaluate the performance of the Ridge regression model on the test data
# The score method returns the R^2 (coefficient of determination) of the model on the test set
rige.score(x_test, y_test)
```
